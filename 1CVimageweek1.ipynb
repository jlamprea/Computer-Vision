```python
import numpy as np
import matplotlib.pyplot as pl
import PIL
```

```python
image_path= "/content/r2d2.jpg" # el archivo se borra del directorio al cerrar el notebook, toca volverlo a cargar al directorio
```

```python
imagen= PIL.Image.open(image_path).convert("L") # ojo width x height en gris 
print(imagen)
```

```python
# convertir la imagen a array numpy
imagen_array= np.asarray(imagen)
```

```python
print(imagen_array.shape) # height x width
```

```python
print(imagen_array)
```

```python
print(imagen_array/255) # normalizar a 0 y 1
```

```python
# dibujar la imagen en gris
pl.imshow(imagen_array,cmap="gray",vmin=0, vmax=255,)
```

```python
imagen_color= PIL.Image.open("/content/r2d2.jpg").convert("RGB")
print(imagen_color)
```

```python
imagen_color_array= np.asanyarray(imagen_color)
print(imagen_color_array.shape)
print(imagen_color_array)
```

```python
# se rellenan con 0 los que no son de color de la capa 
imagen_color_array_red=imagen_color_array.copy()
imagen_color_array_green=imagen_color_array.copy()
imagen_color_array_blue=imagen_color_array.copy()
imagen_color_array_red[:,:,1].fill(0) # llegan de 0 todas las filas y columna 1
imagen_color_array_red[:,:,2].fill(0) # llena de 0 todas las filas y columna 2 queda solo la 0 que es roja
imagen_color_array_green[:,:,0].fill(0)
imagen_color_array_green[:,:,2].fill(0)
imagen_color_array_blue[:,:,0].fill(0)
imagen_color_array_blue[:,:,1].fill(0)
```

```python
# dibujar las imagenes separadas por capas de RGB
pl.figure()
pl.imshow(imagen_color_array_red,vmin=0, vmax=255)
pl.figure()
pl.imshow(imagen_color_array_green,vmin=0, vmax=255)
pl.figure()
pl.imshow(imagen_color_array_blue,vmin=0, vmax=255)
```

```python
import skimage.transform as sk # libreria para cambiar tamaño la imagen
```
